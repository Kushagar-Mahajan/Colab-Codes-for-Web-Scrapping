```
from urllib import request #for downloading data from URL
from bs4 import BeautifulSoup #for souping
import time
import re #importing regular expression
#pointing to URL
url='https://en.wikipedia.org/wiki/Machine_learning'
url1='https://www.mygov.in/covid-19/'

htmldata=request.urlopen(url)#downloads the data
#htmldata.read() #gives the data in html format
soupdata=BeautifulSoup(htmldata,'html5lib')#html parser is used
atagdata=soupdata.findAll('p') #defining tag we need to extract
#print(atagdata)
#now converting data into string format from HTML format
import time
mydata=""
for i in atagdata:
  mydata +=i.text

#  data cleaning
clean_data=re.sub(r'\[[0-9]*\]',' ',mydata)  #  this will remove 0 or more times number appearing in mydata
clean_data=re.sub(r'\s+',' ',clean_data)  # it will remove one or more white spaces with single white space
clean_data=re.sub(r'[^a-zA-Z]',' ',clean_data)  # it will remove single char from starting of the line
clean_data=re.sub(r'\s+',' ',clean_data)  # it will remove one or more white spaces with single white space
print(clean_data)
print(type(clean_data))

#lets deal with string
import time
newdata=clean_data.split()  #split convert string to list

#checking the split function
#for words in newdata:
  #print(words)
  #time.sleep(1)  #will give one second gap


#deal with data by using NLTK
import nltk
#downloading all dictonary for nltk
#nltk.download('all')
#removing stop words
mypuredata=[]
from nltk.corpus import stopwords #stop words are common worlds in nlp
for i in newdata:
  if i.lower() not in stopwords.words('english'):
    mypuredata.append(i)

print(mypuredata)
freq_data=nltk.FreqDist(mypuredata)
freq_data.plot(20)
```
